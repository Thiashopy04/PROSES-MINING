```python
import pandas as pd

# load dataset
df = pd.read_csv('/content/1767518350948_event_log_rumah_sakit_sintetis.csv')

# tampilkan 10 baris pertama
df.head(10)
```

```python
# cek tipe data
df.info()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

# urutkan berdasarkan case_id dan timestamp
df = df.sort_values(['case_id', 'timestamp'])

df.head()
```

```python
jumlah_case = df['case_id'].nunique()
jumlah_event = len(df)
jumlah_aktivitas = df['activity'].nunique()
jumlah_resource = df['resource'].nunique()

print("Jumlah case           :", jumlah_case)
print("Jumlah event          :", jumlah_event)
print("Jumlah aktivitas unik :", jumlah_aktivitas)
print("Jumlah resource unik  :", jumlah_resource)
```
